```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://redplanetscience.com/"

browser.visit(url)
```

```python
html = browser.html
soup = bs(html, 'html.parser')
print(soup.prettify())
```

```python
titles = soup.find_all('div', class_="content_title")
sums = soup.find_all('div', class_="article_teaser_body")

for title in titles:
    print(title.text)
    print("*"*45)
'''
for summary in sums:
    print(sums)
    print("-"*45)   
    '''
```

```python
results = soup.find_all('div', class_="col-md-12")

for result in results:
    # scrape the article header 
    header = result.find('div', class_="content_title").text.strip()
    
    # scrape the article news region
    region = result.find('div', class_="article_teaser_body").text
    
    
    # print article data
    print('**************')
    print(header)
    print('-----------------')
    print(region)
   
```
